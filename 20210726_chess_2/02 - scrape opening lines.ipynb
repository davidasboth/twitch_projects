# Obtain a dataset of ECO classifications for chess openings

Take the content from here: https://www.365chess.com/eco.php and turn it into a DataFrame

```python
import requests
```

```python
res = requests.get("https://www.365chess.com/eco.php")
res.raise_for_status()
```

Let's inspect what comes back

```python
print(res.text)
```

Now use BeautifulSoup to extract the relevant content

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text)
```

```python
all_lines = soup.select("div.line")
len(all_lines)
```

Write a function to extract the name and opening sequence of a single opening

```python
def extract_single_opening_line(line):
    name = line.select("div.opname")[0].text.strip()
    moves = line.select("div.opmoves")[0].text.strip()
    
    line_dict = {
        "name": name,
        "moves": moves
    }
    
    return line_dict
```

Apply to all the contents of the extracted div tags

```python
import pandas as pd

openings = [extract_single_opening_line(line) for line in all_lines]

opening_df = pd.DataFrame(openings)
opening_df.head()
```

Export

```python
opening_df.to_csv("openings.csv", index=False)
```
